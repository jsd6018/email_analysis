```python
import pandas as pd
import numpy as np
import csv
import os,sys
```

```python
df=pd.read_csv("DIRECTED_WEIGHTED.csv")
df=df.iloc[:,1:]
print(df)
```

```python
df2=pd.read_csv("FINALISED_DATA.csv")
print(df2)
```

```python
win=[]
wout=[]
WIN=0
WOUT=0
for i in range(len(df)):
    print("------ ITERATION ",(i+1)," ------")
    s=df.iloc[i,1]
    print(s)
    for j in range(len(df2)):
        if(df2.iloc[j,0]==s):
            print("FROM")
            print("")
            print(df2.iloc[j,:])
            WOUT=WOUT+df2.iloc[j,2]
        if(df2.iloc[j,1]==s):
            print("TO")
            print("")
            print(df2.iloc[j,:])
            WIN=WIN+df2.iloc[j,2]
    win.append(WIN)
    wout.append(WOUT)
    WIN=0
    WOUT=0
```

```python
df['WEIGHTED_ID']=win
df['WEIGHTED_OUT']=wout
print(df)
```
