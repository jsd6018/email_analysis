```python
data = pd.read_csv("../EMAIL_ASSIGNMENT_2/FINALISED_DATA.csv")
```

```python
G = nx.from_pandas_edgelist(data, 'FROM', 'TO',True,nx.Graph())
print("The number of unique nodes: ",len(G))
print("The number of edges: ",G.number_of_edges())
nx.draw_circular(G)
plt.show()
```

```python
'''============================GIRVAN NEWMAN ALGORITHM IMPLEMENTATION ======================================='''
#============================================================================================================
```

```python
communities_generator = community.girvan_newman(G)
```

```python
communities=[]
for c in next(communities_generator):
    print(c)
    c=list(c)
    communities.append(c)
    C=G.subgraph(c)
    nx.draw_circular(C)
    plt.show()
    #=============================
    print("Clustering Coefficient of each node: ",nx.clustering(C))
    print("Clustering Coefficien of the Network: ",nx.average_clustering(C))
    print("Average shortest path length: ",nx.average_shortest_path_length(C))
    print("")
    print("-----------------------")
```

```python
print(len(communities))  #-----DISPLAYING THE NUMEBR OF COMMUNITIES GENERATED USING GIRVAN-NEWMAN ALGORITHM
```

```python
'''

            FUNCTION DEFINITIONS FOR COMPUTING THE INTERNAL AND EXTERNAL DEGREES OF THE NODES
            
            
            STEPS INVOLVED:  FIND COMMUNITY NUMBER FOR EACH NODE
                             FIND THE NEIGHBORS OF THE NODE
                             COMPUTE THE COMMUNITY NUMBER FOR EACH NEIGHBOR
                             THE NUMBER OF NEIGHBORS IN THE SAME COMMUNITY AS THE NODE GIVES INTERNAL DEGREE
                             EXTERNAL DEGREE = TOTAL DEGREE OF A NODE - INTERNAL DEGREE OF THE NODE

'''
```

```python
def find_cno(node):
    C=-1
    for i in range(len(communities)):
        for j in communities[i]:
            if j==node:
                C=(i+1)
                return C
```

```python
def find_same_community_nodes(cno,l):
    X=0
    for i in range(len(l)):
        if find_cno(l[i])==cno:
            X=X+1
    return X

def find_internal_degree(node,cno):
    l=[n for n in G.neighbors(node)]
    print("The list of neighbors is: ",l)
    COUNT = find_same_community_nodes(cno,l)
    return COUNT
```

```python
def find_external_degree(node,int_degree):
    degree = G.degree(node)
    return degree-int_degree
```
